```python
import json
import pandas as pd

# ✅ Load the file from the data/ directory
with open("../data/pollen_2025-07-30_to_2025-08-03.json", "r", encoding="utf-8") as f:
    data_list = json.load(f)

# ✅ Flatten the data
flattened_rows = []

for day_data in data_list:
    region = day_data.get("regionCode", "Unknown")
    fetched_date = day_data.get("fetched_date", "")

    for daily_info in day_data.get("dailyInfo", []):
        date = daily_info.get("date", {})
        forecast_date = f"{date.get('year', '')}-{date.get('month', ''):02}-{date.get('day', ''):02}"

        for pollen in daily_info.get("pollenTypeInfo", []):
            index = pollen.get("indexInfo", {})
            row = {
                "fetched_date": fetched_date,
                "forecast_date": forecast_date,
                "region": region,
                "pollen_code": pollen.get("code"),
                "pollen_name": pollen.get("displayName"),
                "in_season": pollen.get("inSeason"),
                "index_value": index.get("value"),
                "index_category": index.get("category"),
                "index_description": index.get("indexDescription"),
                "recommendations": " | ".join(pollen.get("healthRecommendations", [])) if pollen.get("healthRecommendations") else None
            }
            flattened_rows.append(row)

# ✅ Convert to DataFrame
df = pd.DataFrame(flattened_rows)

# ✅ Display first few rows
df.head()
```
